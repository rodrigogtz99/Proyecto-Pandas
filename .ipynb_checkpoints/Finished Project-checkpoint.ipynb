```python
northern_seasons = {
    'winter': [12, 1, 2],
    'spring': [3, 4, 5],
    'summer': [6, 7, 8],
    'autumn': [9, 10, 11]
}

southern_seasons = {
    'summer': [12, 1, 2],
    'autumn': [3, 4, 5],
    'winter': [6, 7, 8],
    'spring': [9, 10, 11]
}
```

```python
def find_season(month, hemisphere):
    if hemisphere == 'northern':
        for season, months in northern_seasons.items():
            if month in months:
                return season
    elif hemisphere == 'southern':
        for season, months in southern_seasons.items():
            if month in months:
                return season
    return 'unknown'
```

```python
sharks['Season'] = sharks.apply(lambda row: find_season(row['Month'], row['hemisphere']), axis=1)
```

```python
# Step 1: Filter the DataFrame to exclude rows with 'unknown' season
sharks_filtered = sharks[sharks['Season'] != 'unknown']

# Step 2: Group the data by both 'Season' and 'Real_Year' and calculate the size of each group (number of shark attacks)
season_year_counts = sharks_filtered.groupby(['Season', 'Real_Year']).size().reset_index(name='Attack_Count')

# Step 3: Pivot the table to create a summary table without 'unknown' season
season_year_pivot = season_year_counts.pivot_table(index='Real_Year', columns='Season', values='Attack_Count', fill_value=0)

# Step 4: Calculate the average for each season across all the years using mean()
average_by_season = season_year_pivot.mean()

# Step 5: Calculate the overall average across all the years
overall_average = season_year_pivot.mean().mean()

# Now, you have the pivot table without the 'unknown' season, and you can proceed with your analysis.
```
